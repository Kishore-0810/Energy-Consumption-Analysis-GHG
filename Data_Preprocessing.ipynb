```python
# renaming the columns for all  years dataset before concatenating alltogether 
df1.columns = ['Sector', 'Sub Sector', 'Organization', 'Operation',
               'Operation Type', 'Address', 'City', 'Postal Code', 'Total Indoor Space_x',
               'Unit of Measure', 'Weekly Average Hours', 'Annual Flow (M', 'Electricity_Quantity',
               'Electricity_Unit', 'NaturalGas_Quantity', 'NaturalGas_Unit', 'FuelOil12_Quantity',
               'FuelOil12_Unit', 'FuelOil46_Quantity', 'FuelOil46_Unit', 'Propane_Quantity',
               'Propane_Unit', 'Coal_Quantity', 'Coal_Unit', 'Wood_Quantity', 'Wood_Unit',
               'DistrictHeating_Quantity', 'DistrictHeating_Unit', 'DistrictHeating_IsRenewable',
               'DistrictHeating_RenewableEmissionFactor', 'DistrictCooling_Quantity',
               'DistrictCooling_Unit', 'DistrictCooling_IsRenewable',
               'DistrictCooling_RenewableEmissionFactor', 'Number of Portables',
               'Swimming Pool', 'GHG Emissions KG', 'Energy Intensity ekWh_sqft',
               'Energy Intensity ekWh_mega_litre', 'Energy Intensity GJ_m2',
               'Energy Intensity GJ_mega_litre', 'Year']
df4.rename(columns = {'SubSector': 'Sub Sector'}, inplace = True)
df6.rename(columns = {'SubSector': 'Sub Sector'}, inplace = True)
df7.rename(columns = {'SubSector': 'Sub Sector'}, inplace = True)
df8.rename(columns = {'SubSector': 'Sub Sector'}, inplace = True)
df9.rename(columns = {'SubSector': 'Sub Sector', 'Annual Flow (M)': 'Annual Flow (M' }, inplace = True)
```

```python
# concating all to a single dataframe
total_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], axis = 0, ignore_index = True)
# top five rows
total_df.head()
```

```python
# bottom five rows
total_df.tail()
```

```python
# the shape of the dataset is 155124 rows and 42 columns
total_df.shape
```

```python
# columns / features
total_df.columns
```

```python
# missing values percentage for all columns
for i in total_df.columns:
    print(i, "-->", round(total_df[i].isna().sum()/total_df.shape[0]*100, 2))
```

```python
# getting those columns which were having less than 5 percentage missing values
new_df = total_df.loc[:, total_df.isna().sum()/total_df.shape[0]*100 < 5]
new_df
```

```python
# columns
new_df.columns
```

```python
# checking for missing values
new_df.isna().sum()
```

```python
# dropping those missing values rows
new_df.dropna(inplace = True)
```

```python
# after removing missing values
new_df.isna().sum()
```

```python
# checking for duplicated rows
new_df.duplicated().sum()
```

```python
# the shape of the datset after preprocessing
new_df.shape
```

```python
# resetting index
new_df.reset_index(drop = True, inplace = True)
```

```python
# data types
new_df.dtypes
```

```python
new_df.loc[:,"Coal_Quantity"] = new_df.loc[:,"Coal_Quantity"].astype(float)
```

```python
# saving the cleaned dataset to a csv file
new_df.to_csv("Cleaned_Energy_consumption_GHG_Dataset(2012-2020).csv", index = False, encoding = "utf-8-sig")
```
